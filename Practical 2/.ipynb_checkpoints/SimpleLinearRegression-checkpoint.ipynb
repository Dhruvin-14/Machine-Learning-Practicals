```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('E:\\Semester-6\\Machine Learning\\score.csv')    # student's reading hours and score dataset
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
```

```python
for i in X:
    print(i, end = " ")
print()
```

```python
for j in Y:
    print(j, end = " ")
```

```python
# Splitting dataset into training and testing 

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

```python
# Simple Linear regressioon on training dataset

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)    
```

```python
# Visualising the Training set results

plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Reading Hours vs Marks (Training set)')
plt.xlabel('Scores')
plt.ylabel('Hours')
plt.show()
```

```python
# Visualising the Test set results

plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.title('Reading Hours vs Marks (Test set)')
plt.xlabel('Scores')
plt.ylabel('Hours')
plt.show()
```

```python
# Visualising the Test set results

plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Reading Hours vs Marks (Test set)')
plt.xlabel('Scores')
plt.ylabel('Hours')
plt.show()
```

```python
# Predicting the Test set results

Y_pred = regressor.predict(X_test)
```

```python
print(Y_pred)
```

```python
from sklearn.metrics import mean_absolute_error

MAE =  mean_absolute_error(Y_test, Y_pred)
print("Mean Absolute Error : " , MAE)
```

```python
from sklearn.metrics import mean_squared_error

MSE =  mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error : ", MSE)
```

```python
import math

MSE = np.square(np.subtract(Y_test, Y_pred)).mean()
RMSE = math.sqrt(MSE)
print("Root Mean Square Error : ", RMSE)
```

```python
from sklearn.metrics import r2_score

r = r2_score(Y_test, Y_pred)
print(r)
```

```python
import sklearn.metrics as metrics

r2_sk = metrics.r2_score(Y_test, Y_pred)
N = Y_test.shape[0]
p = 3
x = (1-r)
y = (N-1)/ (N-p-1)
adj_rsquared = (1- (x*y))
print(adj_rsquared)
```
