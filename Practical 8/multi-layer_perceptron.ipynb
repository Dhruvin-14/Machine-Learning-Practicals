```python
import pandas as pd
df = pd.read_csv('C:\\Users\\dhruv\\Machine Learning Practicals\\Practical 8\\winequality-red.csv')
df.head(5)
df.head()
```

```python
from sklearn.neural_network import MLPClassifier
y = df['quality']
y.head()
```

```python
X = df.drop('quality', axis=1, inplace=False)
X.head()
X.describe()
```

```python
model = MLPClassifier(solver='sgd', hidden_layer_sizes=(16, 8), random_state=1, learning_rate_init=0.005, learning_rate='adaptive', verbose=True, validation_fraction=0.1, early_stopping=True)
```

```python
model.fit(X,y)
```

```python
model2 = MLPClassifier(solver='sgd', hidden_layer_sizes=(24, 8), random_state=1, learning_rate_init=0.005, learning_rate='adaptive', verbose=True, validation_fraction=0.1, early_stopping=True)
model2.fit(X,y)
```

```python
import matplotlib.pyplot as plt

plt.plot(model.loss_curve_)


```

```python
plt.plot(model2.loss_curve_)
```

```python
model3 = MLPClassifier(solver='sgd', hidden_layer_sizes=(16, 12), random_state=1, learning_rate_init=0.001, learning_rate='invscaling', verbose=True, validation_fraction=0.1, early_stopping=True)
model3.fit(X,y)
```

```python
plt.plot(model3.loss_curve_)
model3.score(X,y)
```
