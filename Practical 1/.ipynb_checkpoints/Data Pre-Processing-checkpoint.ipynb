```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("E:\\Semester-6\\Machine Learning\\titanic_data.csv")
X = dataset.iloc[:, -14:].values  # Independent Variables 
Y = dataset.iloc[:, :-14].values  # Depended Variables 
```

```python
print(X)
```

```python
for i in Y:
    print(i, end=' ')
```

```python
# Taking Care of missing numeric data 
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 2:3])
X[:, 2:3] = imputer.transform(X[:, 2:3])
```

```python
print(X)
```

```python
# Taking care of missing Charactered or string data 
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(X[:, 10:12])
X[:, 10:12] = imputer.transform(X[:, 10:12])
```

```python
print(X)
```

```python
# Encoding Data 
# Independent variable 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1, 6, 7, 8, 9, 10, 11, 12, 13])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
# Spliting Dataset into training and testing set
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
# Feature Scalling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, [0, 2, 3, 4]] = sc.fit_transform(X_train[:, [0, 2, 3, 4]])
X_test[:, [0, 2, 3, 4]] = sc.transform(X_test[:, [0, 2, 3, 4]])
```
